# Step 1:
|Import pacakges

```python
import networkx as nx
import matplotlib.pyplot as plt
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

# Step 2
Initialize graph

```python
graph = nx.Graph()
graph
```

# Step 3
Add nodes

```python
nodes = ["A", "B", "C", "D"]
graph.add_nodes_from(nodes)
graph.nodes()
```

# Step 4
Add edges

```python
edges = [
    ("A", "B"),
    ("A", "C"),
    ("A", "D"),
    ("B", "D"),
    ("C", "D")
]
graph.add_edges_from(edges)
graph.edges()
```

# Step 5
Visualize the graph

```python
pos = graphviz_layout(graph, prog="dot")
node_size = 1000
node_colors = ["blue", "pink", "orange", "yellow"]
nx.draw(graph,
        pos,
        with_labels=True,
        node_size = node_size,
        node_color = node_colors)
plt.show()
```

# Step 6
Centrality calculation

```python
# Degree centrality
cent_degree = nx.degree_centrality(graph)
cent_degree
```

```python
# Between centrality
cent_btn = nx.betweenness_centrality(graph)
cent_btn
```

```python
# Closeness centrality
cent_close = nx.closeness_centrality(graph)
cent_close
```

```python
# Eigenvector centrality
cent_eigen = nx.eigenvector_centrality(graph)
cent_eigen
```

# Step 7
Degree calculation

```python
degree = nx.degree(graph)
degree
```
