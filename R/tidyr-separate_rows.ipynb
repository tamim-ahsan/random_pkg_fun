What does the separate_rows function do?
- It separates collapsed columns into multiple rows based on a separator

```r
# Load library
library(tidyverse)
```

```r
# Create a tibble
tbl <- tibble::tibble(
  Names = c("Sam", "Tom", "Rony"),
  Languages = c("Bengali,English", "Spanish,English,Arabic", "French,Spanish")
  )
tbl
```

```r
# Separate the Languages column into multiple rows
tbl <- tbl |>
  separate_rows(Languages, sep=",")
print(tbl)
```
